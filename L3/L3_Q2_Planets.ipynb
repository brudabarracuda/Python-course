```python
import matplotlib.pyplot as plt
import numpy as np




axis = np.array([0.39, 0.72, 1.00, 1.52, 5.20, 9.54, 19.22, 30.06, 39.48])

Periods = np.array([0.24, 0.62, 1.00, 1.88, 11.86, 29.46, 84.01, 164.8, 248.09])


names = ["Mercury", "Venus", "Earth", "Mars", "Jupiter", "Saturn", 
         "Uranus", "Neptune", "Pluto"]

colors = ['slategrey','darkgoldenrod','royalblue','firebrick','peru','gold','lightsteelblue','blue','darkred']

plt.figure(figsize=(10, 6))

plt.scatter(axis, Periods, s=30, alpha=0.6, c=colors, label='"Planets"')

for k in range(len(colors)):
    plt.text(axis[k], Periods[k]*0.70, f"{names[k]}",
                            rotation=0, va='bottom', ha='center', fontsize=10,
                            color=colors[k])


plt.yscale("log")
plt.xscale("log")
plt.xlabel("Major Semiaxis (A.U.)")
plt.ylabel("Revolution (years)")
plt.title("Our Solar System")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
#Here is a list of book titles (from http://thegreatestbooks.org). Loop through the list and capitalize each word in each title.

titles = ["don quixote", 
          "in search of lost time", 
          "ulysses", 
          "the odyssey", 
          "war and peace", 
          "moby dick", 
          "the divine comedy", 
          "hamlet", 
          "the adventures of huckleberry finn", 
          "the great gatsby"]

for n in range(len(titles)):
    words =  titles[n].split()   # Get single words for each title
    titles[n] = ""
    
    for i in range(len(words)):
        words[i] = words[i].capitalize()   # Capitalize them
        titles[n] = titles[n] +" " + words[i]
        
    
    #print(words)

print(titles)
    
```

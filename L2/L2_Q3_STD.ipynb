```python
import numpy as np
import time
import random
import matplotlib.pyplot as plt


def STD_compute(a):   # COMPUTES THE STANDARD DEVIATION
    start = time.perf_counter()
    a_average = np.sum(a)/len(a)
    squared_sum = np.sum((a-a_average)**2)
    aaaaa = squared_sum/len(a)
    STD = np.sqrt(aaaaa)
    stop = time.perf_counter()
    elapsed = stop - start
    #print(f"The standard deviation of the {len(a):.0f}-entries vector is {STD:.6f}\n")
    #print(f"Execution time for custm STD computation is  {1000000*elapsed:.3f} us\n")
    return STD, elapsed

def STD_numpy(a):   # COMPUTES THE STANDARD DEVIATION
    starto = time.perf_counter()  
    STD = np.std(a)
    stopo = time.perf_counter()
    elapsedo = stopo - starto
    #print(f"The standard deviation of the {len(a):.0f}-entries vector is {STD:.6f}\n")
    #print(f"Execution time for numpy STD computation is  {1000000*elapsedo:.3f} us\n")
    return STD, elapsedo 



lengths = []
times_custom = []
times_numpy = []

n = 0

for _ in range(1000):
    length = random.randint(1, 100_000)
    a = np.random.uniform(0, 100, size=length)

    STD_custom, t_custom = STD_compute(a)
    STD_np, t_numpy = STD_numpy(a)

    lengths.append(length)
    times_custom.append(t_custom * 1e6)  # convert to microseconds
    times_numpy.append(t_numpy * 1e6)

    if n < 5:
       print(f"The STD of the {len(a):.0f}-entries vector is {STD_custom:.6f} with custom function\n")
       print(f"The standard deviation of the {len(a):.0f}-entries vector is {STD_np:.6f} with numpy STD\n")
       n += 1


plt.figure(figsize=(10, 6))

plt.scatter(lengths, times_custom, s=10, alpha=0.6, label="Custom STD")
plt.scatter(lengths, times_numpy, s=10, alpha=0.6, label="NumPy STD")

plt.yscale("log")
plt.xlabel("Vector length")
plt.ylabel("Execution time (s)")
plt.title("STD computation time vs vector length")
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
    

    
```

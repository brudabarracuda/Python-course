```python
import struct
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
import calendar
import os
from ipywidgets import interact, IntSlider, fixed, Checkbox, FloatText, FloatSlider

from IPython.display import clear_output

%matplotlib widget

MonthName2NumberMap = {month: index for index, month in enumerate(calendar.month_abbr) if month}


alpha_lines = {
    "Po-210": [5304.3],
    "Ra-226": [4784],
    "Rn-222": [5489],
    "Po-218": [6002],
    "Po-214": [7687],
}

def TimeConverter(Time) :
    CorrectTime_Hours = Time / (100 * 3600)     # CHECK WITH MASSI, FROM ONLINE DOCUMENTATION LOOKS LIKE 100 IS ADDED BECAUSE DATA ARE STORED IN 10 ms, TIMES SHOULD CHECK OUT
    return CorrectTime_Hours

def ComputeTotActivity(Histo, Time) :
    TotActivity = sum(Histo) / (Time * 3600)    # Bring back to seconds to compute activity
    print(f"Total Counts = {sum(Histo)}")
    return TotActivity
    
def ReadCHNFile(fileName):
  """ 
  Function to read a CHN file - ORTEC binary format.
  The function decodes the information and returns a dictionary with all the relevant information and the vector of the bin counts (Histogram).
  
  The information stored are :
    - MCA, Segment, StartTime_seconds, Real_Time, Live_Time, StartTime_day, StartTime_month, StartTime_year, StartTime_century, StartTime_hour, StartTime_minute, StartTime, Channel_Offset, Channel_Number, Histogram

  ## Inputs
   - filename [str] path of the chn file to open

  ## Outputs
   - RunInformation [dict] all the run information
   - Histogram [ndarray] array with the counts for each bin

  """
  RunInformation = {}
  with open(fileName, mode='rb') as file: # b is important -> binary

    

    fileContentALL = file.read()
    ftype = struct.unpack('h', fileContentALL[0:2])
    RunInformation["MCA"] = struct.unpack('H', fileContentALL[2:4])[0]
    RunInformation["Segment"] = struct.unpack('H', fileContentALL[4:6])
    RunInformation["StartTime_seconds"] = fileContentALL[6:8].decode()

    #RunInformation["Real_Time"] = struct.unpack("l",fileContentALL[8:12])[0]   # by Mattia for Gamma, wrong format for alpha data!  The problem is the "l" option for unpack 
    #RunInformation["Live_Time"] = struct.unpack("l",fileContentALL[12:16])[0]  # by Mattia for Gamma, wrong format for alpha data! The problem is the "l" option for unpack
      


    RunInformation["Real_Time"] = struct.unpack("<I",fileContentALL[8:12])[0]   
    RunInformation["Live_Time"] = struct.unpack("<I",fileContentALL[12:16])[0]
   
    
    RunInformation["StartTime_day"] = fileContentALL[16:18].decode()
    RunInformation["StartTime_month"] =  fileContentALL[18:21].decode()
    RunInformation["StartTime_year"] =  2000+int(fileContentALL[21:23].decode())
    RunInformation["StartTime_century"] = fileContentALL[23:24].decode()
    RunInformation["StartTime_hour"] = fileContentALL[24:26].decode()
    RunInformation["StartTime_minute"] = fileContentALL[26:28].decode()

    RunInformation["StartTime"] = datetime(RunInformation["StartTime_year"],
                                          MonthName2NumberMap[RunInformation["StartTime_month"]], 
                                          int(RunInformation["StartTime_day"]), 
                                          int(RunInformation["StartTime_hour"]), 
                                          int(RunInformation["StartTime_minute"]), 
                                          int(RunInformation["StartTime_seconds"]), 
                                          0)


    RunInformation["Channel_Offset"] = struct.unpack('H', fileContentALL[28:30])[0]
    RunInformation["Channel_Number"] = struct.unpack('H', fileContentALL[30:32])[0]

    #read the spectrum
    Histogram = []
    for ind in range(RunInformation["Channel_Number"]):
      initial_byte = 32+ind*4
      final_byte = initial_byte+4
      Histogram.append(struct.unpack("<I",fileContentALL[initial_byte:final_byte])[0])   
      #Histogram.append(struct.unpack("l",fileContentALL[initial_byte:final_byte])[0])  # by Mattia for Gamma, wrong format for alpha data! The problem is the "l" option for unpack
      

    Histogram = np.array(Histogram)
    RunInformation["Histogram"] = Histogram

    return RunInformation,Histogram


def rebin_spectrum(counts, rebin_factor=1):
    if rebin_factor <= 1:
        return counts
    n = len(counts)
    n_rebin = n // rebin_factor
    counts = counts[:n_rebin * rebin_factor]
    rebinned = counts.reshape(n_rebin, rebin_factor).sum(axis=1)
    return rebinned

def rebin_channels(channels, rebin_factor=1):
    if rebin_factor <= 1:
        return channels
    n_rebin = len(channels) // rebin_factor
    channels = channels[:n_rebin * rebin_factor]
    rebinned_x = channels.reshape(n_rebin, rebin_factor).mean(axis=1)
    return rebinned_x


# TEST FOR CALIBRATION 

def data_comparison(file1, file2, normalize=False, calibrate=False, a=0, b=1, Emin=None, Emax=None, plot_lines=False):
    """
    Plots spectra from two .Chn files with interactive rebinning,
    log-scale toggle, and optional calibration.

    Calibration: E = a + b*x  (x = channel)
    """

    # Read files
    RunInformation1, Histo1 = ReadCHNFile(file1)
    RealT1 = TimeConverter(RunInformation1["Real_Time"])
    LiveT1 = TimeConverter(RunInformation1["Live_Time"])
    Activity1 = ComputeTotActivity(Histo1, LiveT1)

    RunInformation2, Histo2 = ReadCHNFile(file2)
    RealT2 = TimeConverter(RunInformation2["Real_Time"])
    LiveT2 = TimeConverter(RunInformation2["Live_Time"])
    Activity2 = ComputeTotActivity(Histo2, LiveT2)

    # Normalize if requested
    if normalize:
        Histo1 = Histo1 / (LiveT1 * 3600)
        Histo2 = Histo2 / (LiveT2 * 3600)
        print(f"Total Normalized Counts (file1) = {sum(Histo1):.2f}")
        print(f"Total Normalized Counts (file2) = {sum(Histo2):.2f}")

    # Base channel arrays
    x1 = np.arange(RunInformation1["Channel_Number"])
    x2 = np.arange(RunInformation2["Channel_Number"])

    # Apply calibration if requested
    if calibrate:
        x1 = a + b * x1
        x2 = a + b * x2

     # --- Helper to compute counts in calibrated window ---
    def integrate_energy_window(x, hist, Emin, Emax):
        if Emin is None or Emax is None:
            return 0
        mask = (x >= Emin) & (x <= Emax)
        return np.sum(hist[mask])
    
     
    # --- Inner plotting function for widget ---
    def plot_with_options(rebin_factor=1, log_scale=False, Emin_input=None, Emax_input=None):
        
        clear_output(wait=True)

        plt.figure(figsize=(10, 6))

    # Rebin histograms
        H1r = rebin_spectrum(Histo1, rebin_factor)
        H2r = rebin_spectrum(Histo2, rebin_factor)

    # Rebin x-axis
        x1r = rebin_channels(np.arange(RunInformation1["Channel_Number"]), rebin_factor)
        x2r = rebin_channels(np.arange(RunInformation2["Channel_Number"]), rebin_factor)

    # Apply calibration if enabled
        if calibrate:
            x1r = a + b * x1r
            x2r = a + b * x2r

    # ---- Print integral result first so it updates with sliders ----
        if calibrate and Emin_input is not None and Emax_input is not None:
            plt.axvspan(Emin_input, Emax_input, alpha=0.1, color='green', label="ROI Region")
    # Plot
        plt.plot(x1r, H1r, label=f"{os.path.basename(file1)} (rebin={rebin_factor})", lw=1)
        plt.plot(x2r, H2r, label=f"{os.path.basename(file2)} (rebin={rebin_factor})", lw=1)

        xlabel = "Energy (keV)" if calibrate else "Channel Number"
        plt.xlabel(xlabel)
        plt.ylabel("Counts" + (" (normalized)" if normalize else ""))
        plt.title("Comparison of Two .CHN Spectra")

        if log_scale:
            plt.yscale("log")

        plt.grid(True)
        plt.legend()
        plt.show()

          # âœ… Compute & print ROI counts *after* plotting
        if calibrate and Emin_input is not None and Emax_input is not None:
            c1 = integrate_energy_window(x1r, H1r, Emin_input, Emax_input)
            c2 = integrate_energy_window(x2r, H2r, Emin_input, Emax_input)

            print(f"Energy Window: {Emin_input:.1f} â€“ {Emax_input:.1f} keV")
            print(f"Counts in ROI ({os.path.basename(file1)}): {c1*LiveT1*3600:.1f}")
            print(f"Counts in ROI ({os.path.basename(file2)}): {c2*LiveT2*3600:.1f}")
            print(f"Activity in ROI ({os.path.basename(file1)}): {10e3*c1:.3f} mBq")
            print(f"Activity in ROI ({os.path.basename(file2)}): {10e3*c2:.3f} mBq")

        colors = ["red", "blue", "green", "orange", "purple", "brown", "cyan", "magenta"]

        if calibrate and plot_lines:
            i = 0
            for isotope, energies in alpha_lines.items():
                color = colors[i % len(colors)]  # Cycle through colors
                for E in energies:
                    plt.axvline(E, linestyle='--', alpha=0.2, linewidth=1, color=color)
                    plt.text(E, max(max(H1r), max(H2r))*0.8, isotope,
                             rotation=90, va='bottom', ha='right', fontsize=8, color=color)
                i += 1
        

    # --- Interactive controls ---
    interact(
        plot_with_options,
        rebin_factor=IntSlider(min=1, max=64, step=1, value=1, description="Rebin"),
        log_scale=Checkbox(value=False, description="Log scale"),
        #Emin_input=FloatText(value=Emin or 0, description="Emin (keV)"),  # FIXED 
        #Emax_input=FloatText(value=Emax or 20000, description="Emax (keV)")  # FIXED
        Emin_input=FloatSlider(value=Emin or 0, min=0, max=15000, step=1, description="Emin (keV)"),
        Emax_input=FloatSlider(value=Emax or 15000, min=0, max=15000, step=1, description="Emax (keV)")
    )

    # Summary info
    print("âœ… Comparison Summary")
    for i, (f,R,L,A) in enumerate([(file1,RealT1,LiveT1,Activity1),
                                   (file2,RealT2,LiveT2,Activity2)], 1):
        info = RunInformation1 if i==1 else RunInformation2
        print(f"\nFile {i}: {os.path.basename(f)}")
        print(f"MCA # {info['MCA']}")
        print(f"Correct Real time = {R:.3f} h")
        print(f"Correct Live time = {L:.3f} h")
        print(f"Start time = {info['StartTime']}")
        print(f"Channel Offset = {info['Channel_Offset']}")
        print(f"Channel Number = {info['Channel_Number']}")
        print(f"Activity Full Range = {10e3*A:.3f} mBq")

    if calibrate:    
        print(f"\nðŸ“ Calibration applied: E = {a} + {b}Â·Channel")


    


if __name__ == "__main__":

  fileName = "/home/brada/Desktop/CCVR3 Stuff/Alpha Spectroscopy G828-2/CAN2_Cupid_LMO-Siccas_G8382_4_24-10-25.Chn"

  '''

  RunInformation,Histogram = ReadCHNFile(fileName)
  RealT = TimeConverter(RunInformation["Real_Time"])
  LiveT = TimeConverter(RunInformation["Live_Time"])
  print()
  print(f'MCA # {RunInformation["MCA"]}')
  #print(f'Real time = {RunInformation["Real_Time"]:d}')
  #print(f'Live time = {RunInformation["Live_Time"]:d}')
    
  print(f'Correct Real time = {RealT:.3f} h')
  print(f'Correct Live time = {LiveT:.3f} h')
  
  print(f'Start time = { RunInformation["StartTime"]}')
  print(f'Channel Offset = {RunInformation["Channel_Offset"]:d}')
  print(f'Channel Number = {RunInformation["Channel_Number"]:d}')


  # %%
  fig, ax = plt.subplots()
  ax.step(np.arange(RunInformation["Channel_Number"]),Histogram)

  '''

    
  file1 = "/home/brada/Desktop/CCVR3 Stuff/Alpha Spectroscopy G828-2/CAN2_Cupid_LMO-Siccas_G8382_4_24-10-25.Chn"
  file2 = "/home/brada/Desktop/CCVR3 Stuff/Alpha Spectroscopy G828-2/Can2_BKG_PiastRamePlasmaLug25-H2_05-12-25.Chn"
    
  #file2 = "/home/brada/Desktop/CCVR3 Stuff/Alpha Spectroscopy G828-2/Can2_BKG_PiastRamePlasmaLug25_01-10-25.Chn"  # OLD BKG CON CAMERATTA COPETA DALLA PIASTRINA

  #data_comparison(file1, file2, normalize=True)
  #data_comparison(file1, file2, normalize=False)

  data_comparison(file1, file2, normalize=True, calibrate=True, a=327.72,b=0.700, Emin=3600, Emax=10000, plot_lines=True)
    

```

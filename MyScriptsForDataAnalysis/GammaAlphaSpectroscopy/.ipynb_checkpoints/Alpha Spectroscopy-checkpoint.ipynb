```python
import struct
import numpy as np
import matplotlib.pyplot as plt
import os

def read_maestro_chn(filename):
    """
    Reads an ORTEC MAESTRO .Chn file and returns:
    - channels (np.array)
    - counts (np.array)
    - metadata (dict)
    """
    with open(filename, "rb") as f:
        version = struct.unpack("<H", f.read(2))[0]
        num_channels = struct.unpack("<H", f.read(2))[0]
        live_time = struct.unpack("<f", f.read(4))[0]
        real_time = struct.unpack("<f", f.read(4))[0]

        data = f.read()
        counts = np.frombuffer(data, dtype=np.uint32)

        # Fallback if 16-bit integers were used
        if len(counts) != num_channels:
            counts = np.frombuffer(data, dtype=np.uint16)

        counts = counts[:num_channels]

    channels = np.arange(num_channels)
    metadata = {
        "version": version,
        "num_channels": num_channels,
        "live_time": live_time,
        "real_time": real_time,
    }
    return channels, counts, metadata


def plot_two_chn(file1, file2, normalize=False):
    """
    Plots spectra from two .Chn files on the same figure.
    Optionally normalizes both spectra by live time.
    """
    channels1, counts1, meta1 = read_maestro_chn(file1)
    channels2, counts2, meta2 = read_maestro_chn(file2)

    # Normalize if requested
    if normalize:
        counts1 = counts1 / meta1["live_time"]
        counts2 = counts2 / meta2["live_time"]

    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(channels1, counts1, label=f"{os.path.basename(file1)}", lw=1)
    plt.plot(channels2, counts2, label=f"{os.path.basename(file2)}", lw=1)
    plt.title("Comparison of Two .Chn Spectra")
    plt.xlabel("Channel Number")
    plt.ylabel("Counts" + (" (normalized)" if normalize else ""))
    plt.legend()
    plt.grid(True)
    plt.show()

    # Console summary
    print("âœ… Comparison Summary")
    print(f"File 1: {file1}")
    print(f"  Channels: {meta1['num_channels']}, Total Counts: {counts1.sum():.0f}")
    print(f"  Live Time: {meta1['live_time']:.2f} s, Real Time: {meta1['real_time']:.2f} s")
    print(f"File 2: {file2}")
    print(f"  Channels: {meta2['num_channels']}, Total Counts: {counts2.sum():.0f}")
    print(f"  Live Time: {meta2['live_time']:.2f} s, Real Time: {meta2['real_time']:.2f} s")


```

```python
import struct
import numpy as np
import matplotlib.pyplot as plt
import os
from collections import namedtuple

Detection = namedtuple("Detection", ["num_channels", "live_time", "real_time",
                                     "header_size", "counts_dtype", "counts_len"])

def try_unpack(fmt, b):
    try:
        return struct.unpack(fmt, b)[0]
    except struct.error:
        return None

def detect_header_and_counts(filename):
    """
    Try to detect:
     - num_channels (uint16 or uint32 at a few offsets)
     - live_time, real_time (float32/float64 at various offsets)
     - header size (common sizes tried)
     - counts dtype (uint32 or uint16)
    Returns a Detection instance (best guess) or None if no match found.
    """
    with open(filename, "rb") as f:
        data_all = f.read()
    filesize = len(data_all)
    # Read first 512 bytes for inspection
    header_bytes = data_all[:512] if filesize >= 512 else data_all

    # Candidate offsets and formats
    num_ch_offsets = [2, 4, 6, 8, 10]  # common places to find channel count
    time_offsets = list(range(4, 64, 4))  # search offsets for floats/doubles
    header_sizes = [12, 32, 64, 128, 256, 512]
    counts_dtypes = [("uint32", np.uint32, 4), ("uint16", np.uint16, 2)]

    candidates = []

    # Try uint16 and uint32 for num_channels at candidate offsets
    for off in num_ch_offsets:
        # uint16
        if off + 2 <= len(header_bytes):
            num_ch = try_unpack("<H", header_bytes[off:off+2])
            if num_ch is not None:
                # sanity check
                if 8 <= num_ch <= 16384:
                    # now try to find live/real time near header
                    for toff in time_offsets:
                        # float32
                        if toff + 4 <= len(header_bytes):
                            lt = try_unpack("<f", header_bytes[toff:toff+4])
                            rt = try_unpack("<f", header_bytes[toff+4:toff+8]) if (toff+8 <= len(header_bytes)) else None
                            # sanity check times (non-negative, not huge)
                            if lt is not None and rt is not None and (0 <= lt < 1e7) and (0 <= rt < 1e7):
                                # try header sizes to see which yields correct counts length
                                for hsize in header_sizes:
                                    if filesize >= hsize + num_ch * 2:
                                        # check uint16
                                        remaining = data_all[hsize:]
                                        counts16_len = len(remaining) // 2
                                        counts32_len = len(remaining) // 4
                                        # prefer uint32 if exact match, else uint16
                                        if counts32_len >= num_ch and (counts32_len == num_ch or counts32_len >= num_ch):
                                            # check interpreting as uint32 yields at least num_ch entries
                                            candidates.append(Detection(num_channels=num_ch, live_time=float(lt),
                                                                        real_time=float(rt), header_size=hsize,
                                                                        counts_dtype="uint32", counts_len=counts32_len))
                                        if counts16_len >= num_ch:
                                            candidates.append(Detection(num_channels=num_ch, live_time=float(lt),
                                                                        real_time=float(rt), header_size=hsize,
                                                                        counts_dtype="uint16", counts_len=counts16_len))
                        # float64
                        if toff + 8 <= len(header_bytes):
                            lt_d = try_unpack("<d", header_bytes[toff:toff+8])
                            rt_d = try_unpack("<d", header_bytes[toff+8:toff+16]) if (toff+16 <= len(header_bytes)) else None
                            if lt_d is not None and rt_d is not None and (0 <= lt_d < 1e7) and (0 <= rt_d < 1e7):
                                for hsize in header_sizes:
                                    if filesize >= hsize + num_ch * 2:
                                        remaining = data_all[hsize:]
                                        counts16_len = len(remaining) // 2
                                        counts32_len = len(remaining) // 4
                                        if counts32_len >= num_ch:
                                            candidates.append(Detection(num_channels=num_ch, live_time=float(lt_d),
                                                                        real_time=float(rt_d), header_size=hsize,
                                                                        counts_dtype="uint32", counts_len=counts32_len))
                                        if counts16_len >= num_ch:
                                            candidates.append(Detection(num_channels=num_ch, live_time=float(lt_d),
                                                                        real_time=float(rt_d), header_size=hsize,
                                                                        counts_dtype="uint16", counts_len=counts16_len))

        # uint32 for channel count
        if off + 4 <= len(header_bytes):
            num_ch32 = try_unpack("<I", header_bytes[off:off+4])
            if num_ch32 is not None and 8 <= num_ch32 <= 16384:
                for toff in time_offsets:
                    if toff + 4 <= len(header_bytes):
                        lt = try_unpack("<f", header_bytes[toff:toff+4])
                        rt = try_unpack("<f", header_bytes[toff+4:toff+8]) if (toff+8 <= len(header_bytes)) else None
                        if lt is not None and rt is not None and (0 <= lt < 1e7) and (0 <= rt < 1e7):
                            for hsize in header_sizes:
                                if filesize >= hsize + num_ch32 * 2:
                                    remaining = data_all[hsize:]
                                    counts16_len = len(remaining) // 2
                                    counts32_len = len(remaining) // 4
                                    if counts32_len >= num_ch32:
                                        candidates.append(Detection(num_channels=num_ch32, live_time=float(lt),
                                                                    real_time=float(rt), header_size=hsize,
                                                                    counts_dtype="uint32", counts_len=counts32_len))
                                    if counts16_len >= num_ch32:
                                        candidates.append(Detection(num_channels=num_ch32, live_time=float(lt),
                                                                    real_time=float(rt), header_size=hsize,
                                                                    counts_dtype="uint16", counts_len=counts16_len))
                    if toff + 8 <= len(header_bytes):
                        lt_d = try_unpack("<d", header_bytes[toff:toff+8])
                        rt_d = try_unpack("<d", header_bytes[toff+8:toff+16]) if (toff+16 <= len(header_bytes)) else None
                        if lt_d is not None and rt_d is not None and (0 <= lt_d < 1e7) and (0 <= rt_d < 1e7):
                            for hsize in header_sizes:
                                if filesize >= hsize + num_ch32 * 2:
                                    remaining = data_all[hsize:]
                                    counts16_len = len(remaining) // 2
                                    counts32_len = len(remaining) // 4
                                    if counts32_len >= num_ch32:
                                        candidates.append(Detection(num_channels=num_ch32, live_time=float(lt_d),
                                                                    real_time=float(rt_d), header_size=hsize,
                                                                    counts_dtype="uint32", counts_len=counts32_len))
                                    if counts16_len >= num_ch32:
                                        candidates.append(Detection(num_channels=num_ch32, live_time=float(lt_d),
                                                                    real_time=float(rt_d), header_size=hsize,
                                                                    counts_dtype="uint16", counts_len=counts16_len))

    # Pick best candidate: prefer exact counts_len == num_channels and uint32 over uint16, larger header_size preference = safer
    if not candidates:
        return None

    def score(c: Detection):
        score = 0
        # prefer exact match
        if c.counts_dtype == "uint32":
            score += 5
        # prefer counts_len close to num_channels
        score -= abs(c.counts_len - c.num_channels) / max(1, c.num_channels)
        # prefer larger header size (more likely real)
        score += (c.header_size // 8)
        # non-zero live_time
        if c.live_time > 0:
            score += 2
        return score

    best = max(candidates, key=score)
    return best

def read_maestro_chn_autodetect(filename):
    """
    Read a MAESTRO .Chn file with autodetection of header and dtype.
    Returns channels (np.array), counts (np.array), metadata (dict)
    """
    detect = detect_header_and_counts(filename)
    if detect is None:
        raise ValueError(f"Could not autodetect header/data layout for file: {filename}. "
                         "Try printing the header hex for manual inspection.")

    with open(filename, "rb") as f:
        all_bytes = f.read()

    header_size = detect.header_size
    num_ch = detect.num_channels

    # interpret counts
    offset = header_size
    if detect.counts_dtype == "uint32":
        counts = np.frombuffer(all_bytes[offset:offset + num_ch * 4], dtype=np.uint32)
    else:
        counts = np.frombuffer(all_bytes[offset:offset + num_ch * 2], dtype=np.uint16)

    # safety: ensure counts length exactly num_ch (truncate/pad)
    counts = counts[:num_ch]
    if counts.size < num_ch:
        # pad with zeros if file shorter than header indicates
        counts = np.pad(counts, (0, num_ch - counts.size), mode='constant', constant_values=0)

    channels = np.arange(num_ch)
    metadata = {
        "num_channels": num_ch,
        "live_time": detect.live_time,
        "real_time": detect.real_time,
        "header_size": header_size,
        "counts_dtype": detect.counts_dtype
    }
    return channels, counts, metadata

def inspect_header_hex(filename, n=128):
    with open(filename, "rb") as f:
        b = f.read(n)
    return b.hex()

def plot_two_chn_autodetect(file1, file2, normalize=False):
    c1, counts1, m1 = read_maestro_chn_autodetect(file1)
    c2, counts2, m2 = read_maestro_chn_autodetect(file2)

    if normalize:
        # normalize by live_time if positive
        if m1["live_time"] > 0:
            counts1 = counts1 / m1["live_time"]
        if m2["live_time"] > 0:
            counts2 = counts2 / m2["live_time"]

    plt.figure(figsize=(10, 6))
    plt.plot(c1, counts1, label=f"{os.path.basename(file1)} (dtype={m1['counts_dtype']}, hdr={m1['header_size']})", lw=1)
    plt.plot(c2, counts2, label=f"{os.path.basename(file2)} (dtype={m2['counts_dtype']}, hdr={m2['header_size']})", lw=1)
    plt.title("Comparison of Two .Chn Spectra (autodetect)")
    plt.xlabel("Channel Number")
    plt.ylabel("Counts" + (" (normalized by live time)" if normalize else ""))
    plt.legend()
    plt.grid(True)
    plt.show()

    # summary
    print("File 1:", file1)
    print(m1)
    print("Total counts:", counts1.sum())
    print()
    print("File 2:", file2)
    print(m2)
    print("Total counts:", counts2.sum())

if __name__ == "__main__":
    # Example usage â€” replace with your actual filenames
    f1 = "/home/brada/Desktop/CCVR3 Stuff/Alpha Spectroscopy G828-2/Can2_BKG_PiastRamePlasmaLug25_01-10-25.Chn"
    f2 = "/home/brada/Desktop/CCVR3 Stuff/Alpha Spectroscopy G828-2/CAN2_Cupid_LMO-Siccas_G8382_4_24-10-25.Chn"

    try:
        plot_two_chn_autodetect(f1, f2, normalize=False)
    except Exception as e:
        print("Error:", e)
        # helpful debug: print header hex
        print("Header hex (first 128 bytes) of file1:")
        print(inspect_header_hex(f1))
        print("Header hex (first 128 bytes) of file2:")
        print(inspect_header_hex(f2))
```
